###  Validação de entrada para login

João está desenvolvendo um sistema de cadastro para um site de leitura. Ele precisa garantir que os usuários insiram um nome de usuário e uma senha válidos. As regras são as seguintes:

O nome de usuário deve ter pelo menos 5 caracteres.
A senha deve ter pelo menos 8 caracteres.
João quer que o sistema continue solicitando as informações até que ambas as condições sejam atendidas. Quando o usuário insere dados válidos, o programa deve exibir a mensagem: "Cadastro realizado com sucesso!".

Crie um programa que implemente essa lógica usando um laço while.

Saída esperada:

image.png

```python
def login():
    nome_valido = False
    senha_valida = False

    while not nome_valido or not senha_valida:
      if nome_valido == False:
        login_str = input('Digite seu nome de usuário: ')
        if len(login_str) >= 5:
          nome_valido = True
        else:
          print('O nome de usuário deve conter pelo menos 5 caracteres.')
          nome_valido = False
          continue

      senha_str = input('Digite sua senha: ')
      if len(senha_str) >= 8:
        senha_valida = True
      else:
        print('A senha deve conter pelo menos 8 caracteres.')
        senha_valida = False

    print("Cadastro realizado com sucesso!")

login()
```

```python
while True:
    nome_usuario = input("Digite seu nome de usuário: ")
    senha = input("Digite sua senha: ")

    if len(nome_usuario) < 5:
        print("O nome de usuário deve ter pelo menos 5 caracteres.")
        continue

    if len(senha) < 8:
        print("A senha deve ter pelo menos 8 caracteres.")
        continue

    print("Cadastro realizado com sucesso!")
    break
```

Dado o seguinte código com uma lista de números de telefone armazenados incorretamente como str, faça duas funções, uma que converte os tipos para inteiro e outra que verifica se a conversão foi feita corretamente e todos os números de telefone são inteiros:

Exemplo de entrada:

```python
telefones = ["11987654321", "21912345678", "31987654321", "11911223344"]
```

Saida esperada:


Todos os números foram convertidos corretamente!

```python
print(type(telefones[1]))
```

```python
def transforma_int(lista): # cria a funçao e recebe lista como arguimento
    lista_inteiros = [int(item) for item in lista] # cria uma variavel que é igual a int(item) para item na lista
    return lista_inteiros #retorna a lista


telefones = transforma_int(telefones) # muda a variavel chamando a funçao a cima para transformala

def checa_int(lista): #cria outra função qual tbm recebe uma lista como argumento
  resultado = all(isinstance(x, type(lista[0])) for x in lista) # define a variavel resultado igual verifica se cada 'x' tem o mesmo tipo que o primeiro item lista[0]
  if resultado is True and type(telefones[1]) is int: # se o resultado for true indicando que todos os dados sao do mesmo tipo e que eles sao numerros inteiros
    print('Todos os números foram convertidos corretamente!') # imprime a mensagem que deu certo
  else: # se nao
    print('Os números não foram convertidos.') # imprime que nao deu

checa_int(telefones) # chama a funçao
```
