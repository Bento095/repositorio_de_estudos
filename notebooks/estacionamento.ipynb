# Rascunho do estacionamento

```python
import math

''' rascunho de um calculador de preço para estacionamento'''

valor_entrada = 6.00
valor_minuto = 2 #apos 40 min cada 10 min custa 2 reais ate o teto, começamdo de 6 reais

hora_de_entrada_str = input('Digite a hora de entrada (HH.MM): ')
hora_de_saida_str = input('Digite a hora de saída (HH.MM): ')
veiculo = int(input('Digite 1 para carro e 2 para caminhonete: '))
modelo_cor_placa = input('Digite o modelo, cor e placa do veículo, separado por espaço: ').split()

if veiculo == 1:
    valor_entrada = 6.00
else:
    valor_entrada = 8.00
    valor_minuto = 2.60

def converter_para_minutos_totais(hora_str):
    if '.' in hora_str:
        horas, minutos = map(int, hora_str.split('.'))
        return horas * 60 + minutos
    else:
        horas = int(hora_str)
        return horas * 60

def calcular_custo(hora_de_entrada_str, hora_de_saida_str):
    hora_entrada_total_minutos = converter_para_minutos_totais(hora_de_entrada_str)
    hora_saida_total_minutos = converter_para_minutos_totais(hora_de_saida_str)

    tempo_de_estacionamento_em_minutos = hora_saida_total_minutos - hora_entrada_total_minutos

    total = valor_entrada

    if tempo_de_estacionamento_em_minutos <= 39:
        pass # Nenhuma cobrança extra para os primeiros 39 minutos
    elif tempo_de_estacionamento_em_minutos >= 180 and veiculo == 1:
        total = 35
    elif tempo_de_estacionamento_em_minutos >= 180:
        total = 45
    else:
        tempo_extra_em_minutos = tempo_de_estacionamento_em_minutos - 39

        # Cada bloco de 10 minutos custa valor_minuto
        extra_cost = math.ceil(tempo_extra_em_minutos / 10) * valor_minuto
        total += extra_cost

    print(f'Cliente {modelo_cor_placa} ficou {tempo_de_estacionamento_em_minutos:.2f} minutos, valor R${total:.2f}')
    return total

valor = calcular_custo(hora_de_entrada_str, hora_de_saida_str)
```
